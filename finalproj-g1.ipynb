```python
import pandas
import sklearn
import numpy

# import hazm

import time
import gc
```

```python
raw_train_data = pandas.read_csv('../input/computational-intelligence-course-final-project/train.csv')
```

```python
raw_test_data = pandas.read_csv('../input/computational-intelligence-course-final-project/test.csv')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def PreProcess(X):
  vectorizer = TfidfVectorizer()

  df = pandas.DataFrame.sparse.from_spmatrix(vectorizer.fit_transform(X))

  return df, vectorizer.get_feature_names()
```

```python
X, dictionary = PreProcess(raw_train_data['Text'])
Y = raw_train_data['Category']
```

```python
del raw_train_data
gc.collect()
```

```python
from sklearn import linear_model
from sklearn import neighbors
from sklearn import svm

# model = linear_model.SGDClassifier()
# model = neighbors.KNeighborsClassifier(neighbors_num)
model = svm.LinearSVC(C=0.1)
# model = svm.NuSVC(gamma='auto')
# model = svm.SVC()
```

```python
model.fit(X, Y)
```

```python
def test_preprocess(T, dictionary):
  vectorizer = TfidfVectorizer(vocabulary = dictionary)
  df = pandas.DataFrame.sparse.from_spmatrix(vectorizer.fit_transform(T))

  return df
```

```python
test_preprocessed = test_preprocess(raw_test_data['Text'], dictionary)
```

```python
prediction = model.predict(test_preprocessed)
```

```python
result = pandas.DataFrame({'Id' : range(len(test_preprocessed)), 'Category' : prediction})
result.to_csv('./result.csv', index=False)
```
